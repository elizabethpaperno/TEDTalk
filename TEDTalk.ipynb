### Plotting with Matplotlib

```python
temp = df["Temperature (C)"]

plt.scatter(pd.to_datetime(df["Formatted Date"],utc=True), temp)
plt.title("Weather")
plt.xlabel("Year")
plt.ylabel("Temp (C)")
plt.show()

plt.boxplot(temp, vert=False)
plt.show()

temp_mean = temp.mean()
temp_median = temp.median()

print('Measures of Central Tendency (MoCT)')
print('mean: ', temp_mean)
print('median: ', temp_median)

print("\n")

temp_range = temp.max() - temp.min()
temp_iqr = scipy.stats.iqr(temp)
temp_std = temp.std()

print('Measures of Variability (MoV):')
print('range: ', temp_range)
print('interquartile range: ', temp_iqr)
print('standard deviation: ', temp_std)
```

## Groupby and Heatmaps
**Heatmaps** a are two-dimensional representation of data in which values are represented by colors. This allows the viewer of the data to have an easy visual to interpert data.Generally there are two axes that map to a value, and the magnitude of this value is represented by **varying colors**.

**Seaborn** (sns) is generally the library that is used for creating heatmaps in python. 

```python
#creates two new cols with month and year derieved from the datetime format
df['Date'] = pd.to_datetime(df["Formatted Date"],utc=True)
df['Year'] = df['Date'].dt.year 
df['Month'] = df['Date'].dt.month 

#only includes needed columns
filtered_df = df.loc[:, ['Month', 'Year', 'Temperature (C)']]

groupedMnYr = filtered_df.groupby(['Month', 'Year'], as_index=False).mean() 

groupedMnYr = groupedMnYr.pivot(index='Month', columns='Year', values='Temperature (C)')
display(groupedMnYr)
sns.heatmap(groupedMnYr)
```

### LSTM Model

```python
# Import the libraries
print("hi")
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
plt.style.use('fivethirtyeight')
print("bye")
```

```python
import yfinance as yf
df = yf.download(tickers="MSFT", period="20d", interval="15m", auto_adjust=True)
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(16,8))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Closing Price USD($)', fontsize=18)
plt.show()
```

```python
# Create a new dataframe with only the 'Close' column
data = df.filter(['Close'])
# Convert the datafram to a numpy array
dataset = data.values
# Get the number of rows to train the model on
training_data_len = math.ceil(len(dataset) * .8) # use about 80% of data for trainng set

training_data_len
```

```python
scaler = MinMaxScaler(feature_range=(0,1)) # sets data to values b/w 0 and 1(inclusive)
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
train_data = scaled_data[0:training_data_len, :]
# Split the data into x_train & y_train data sets
x_train = []
y_train = []

for i in range(10, len(train_data)):
  x_train.append(train_data[i-10:i, 0])
  y_train.append(train_data[i, 0])
  if i <= 11:
    print(x_train)
    print(y_train)
```

```python
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], 1, x_train.shape[1]))
x_train.shape
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(1,x_train.shape[2])))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', run_eagerly=True)
```

```python
# Train the model
model.fit(x_train, y_train, batch_size=30, epochs=10) # epoch is the number of iteratons when your entire data set is passed through forward and backward through a neural network
```

```python
test_data = scaled_data[training_data_len - 10 : 521, :]
# Split and create x_test & y_test
x_test = []
y_test = dataset[training_data_len:, :]
for i in range(10, len(test_data)):
  x_test.append(test_data[i-10:i, 0])

x_test = np.array(x_test)
print(x_test)
```

```python
x_test.shape
```

```python
x_test = np.reshape(x_test, (x_test.shape[0], 1, x_test.shape[1]))
x_test.shape
```

```python
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
```

```python
# Plot the data
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
# Visualize the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Data', fontsize=18)
plt.ylabel('Closing Price USD($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

```python
es = EarlyStopping(monitor = 'val_loss',mode = "min",verbose = 1,patience = 20)
history = model.fit(x_train, y_train, epochs=100, validation_split = 0.2, verbose = 1, callbacks = [es])

plt.plot(history.history['loss'],label = "Training loss")
plt.plot(history.history['val_loss'],label = "Validation loss")
plt.legend()
```
